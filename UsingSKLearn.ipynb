```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 5, 10)

beta_true = np.random.normal(0, 1, 2);
x_tilde = np.hstack((np.ones((len(x), 1)), x.reshape(-1, 1)))

noise = np.random.normal(0, 0.3, len(x))

targets = np.dot(x_tilde, beta_true) + noise


plt.scatter(x, targets)
plt.plot(x, np.dot(x_tilde, beta_true))
plt.show()
```

```python
from sklearn import linear_model

my_regression_model = linear_model.LinearRegression()
my_regression_model.fit(x_tilde, targets)

x_test = np.linspace(0, 5, 100)
x_test = np.hstack((x_test.reshape(-1, 1), x_test.reshape(-1, 1)))
prediction = my_regression_model.predict(x_test)


plt.scatter(x, targets, color="red")
plt.plot(x, np.dot(x_tilde, beta_true))
plt.plot(x_test, prediction, color="g")
plt.show()
```
