```python
import numpy as np
from matplotlib import pyplot as plt
import time
```

```python
x = np.linspace(0, 5, 10)
# Array of 1 and then x values, to multiply 
x_tilde = np.hstack((np.ones((len(x), 1)).reshape(-1, 1), x.reshape(-1, 1)))

# print(x_tilde)

beta0 = 0
beta1 = 1
learning_rate = 0.01

beta = np.array([beta0, beta1])

targets = np.dot(x_tilde, beta)
t_noisy = targets + np.random.normal(0, 1, len(x))

beta_init = np.random.normal(0, 1, 2)
beta = beta_init

for iter in range(100):
    gradient_b0 = np.sum(t_noisy - np.dot(x_tilde, beta))*(-1)
    gradient_b1_temp = np.sum(t_noisy - np.dot(x_tilde, beta))*(1)
    gradient_b1 = np.sum(np.multiply(gradient_b1_temp, x))
    
    loss = 0
    for i in range(len(x)):
        loss += (t_noisy[i] - np.dot(x_tilde, beta)[i])**2
    loss *= np.true_divide(1, 2*len(x))
    print(loss)
    
    beta[0] = beta[0] - learning_rate * gradient_b0
    beta[1] = beta[1] - learning_rate * gradient_b1
    
#     targets = np.dot(x_tilde, beta)
    
# print(t_noisy)
plt.plot(x, targets, "g")
plt.scatter(x, t_noisy, color="red")
plt.show()
#     plt.close()

print(beta)
```
